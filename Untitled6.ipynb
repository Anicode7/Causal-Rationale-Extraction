```python
# Prompt the user for the domain and intent to filter
# filtered_domain = input("#Enter the domain to filter by (e.g., 'Hotel', 'Flight'):")
filtered_domain = "Retail"
# filtered_intent = input("#Enter the intent to filter by (e.g., 'Refund', 'Cancellation'): ")
filtered_intent = "Product Returns"

# Filter the DataFrame
try:    
    filtered_df = df[(df['domain'] == filtered_domain) & (df['intent'] == filtered_intent)]
    print(f"\nFiltered data for Domain: '{filtered_domain}' and Intent: '{filtered_intent}'")
    display(filtered_df.head())
    print(f"Total records after filtering: {len(filtered_df)}")
except KeyError as e:
    print(f"Error: Column '{e}' not found in the DataFrame. Please ensure 'domain' and 'intent' columns exist.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")  
```

```python
import faiss
import torch
from sentence_transformers import SentenceTransformer
import pickle

# 1. Load passages (list of dicts: {"id": str, "text": str})
# Using the filtered_df as the source for passages
passages = filtered_df.to_dict('records')

# Extract all conversational texts from the 'conversation' list within each passage
all_conversation_texts = []
for p_dict in passages:
    for utterance in p_dict['conversation']:
        all_conversation_texts.append(utterance['text'])
texts = all_conversation_texts

device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)
# 2. Load embedding model (bge-large-en-v1.5)
embed_model = SentenceTransformer("BAAI/bge-large-en-v1.5", device=device)

# 3. Encode passages
embs = embed_model.encode(texts, batch_size=32, convert_to_numpy=True, show_progress_bar=True, normalize_embeddings=True)

# 4. Build FAISS IndexFlatIP
d = embs.shape[1]
index = faiss.IndexFlatIP(d)
index.add(embs)

# 5. Save index and metadata
faiss.write_index(index, "passages.index")
with open("passages_meta.pkl", "wb") as f:
    pickle.dump(passages, f)
```

```python
# -----------------------------
# Question decomposition module
# -----------------------------
DECOMPOSE_PROMPT = """You are a Question Decomposer specialized for multi-cluster retrieval.

Input: a single natural-language user query (about an event, complaint, request, or pattern) plus optional context metadata (user_profile, timeframe, transcript_id).

Task: rewrite the input into the cluster-specific retrieval queries — one for each satisfaction label: {-1,0,1} — that a retriever can execute to extract factual passages relevant to the original question.

Output format rules (strict):
- Produce exactly 3 lines, one line per label.
- Each line must begin with the Label name followed by a colon and then the cluster-specific query and retrieval_conditions JSON block.
- No numbering, no explanation, no additional text.
- MUST OUTPUT THESE LINES (in this order): -1,0,1.
- Example line format:
  Hotel: "rewritten query text" {"keywords":["k1","k2"],"fuzzy":true,"temporal_anchor":"before|after|none","priority":"high|medium|low"}
-After producing exactly six domain lines in the required order, append one final line containing exactly this JSON and nothing else: {"lines":6}

Rewriting rules:
- Transform the generic query so each Cluster-specific line asks for the precise factual evidence that would answer the original query *in that cluster* (e.g., for refunds/policies use policy-terms; for telecom use plan/coverage; for flight use booking/cancellation; for hotel use reservation/room-charge; for retail use transaction/return; for banking use charge/transaction dispute).
- Include a concise set of retrieval keywords (3–6) covering synonyms and common ASR noise; set "fuzzy" true for user-spoken text.
- Add metadata_filters when helpful (domain-specific fields like booking_id, policy_type, transaction_code, call_outcome, timeframe). If no metadata is available, use {}.
- Set approx_span_tokens to a sensible range for that domain (typical 10–200 tokens).
- Use temporal_anchor when the information should be searched relative to an anchor event (e.g., "after promise", "before cancellation", "none").
- Mark "priority" high for the most directly relevant sub-query, medium/low for supporting evidence.
- Always ensure that there is an output for all 3 domains.
Hard constraints:
- Always output exactly three Label lines .
- Always add all three labels for every query.
- Keep each rewritten query ≤ 30 tokens.
- Do NOT answer the query — only produce the domain queries + retrieval conditions.

Example:
Query: Why are refunds occuring in Hotel domain?
-1: "evidence of refund causes in hotel interactions with dissatisfied callers" {"keywords":["refund","reimburse","charge issue","booking error","delay"],"fuzzy":true,"metadata_filters":{"domain":"Hotel"},"approx_span_tokens":120,"temporal_anchor":"none","priority":"high"}
0: "neutral-caller explanations for hotel refunds and their triggering issues" {"keywords":["refund","policy","adjustment","billing","promo"],"fuzzy":true,"metadata_filters":{"domain":"Hotel"},"approx_span_tokens":100,"temporal_anchor":"none","priority":"medium"}
1: "satisfied-caller references describing reasons refunds were issued in hotel cases" {"keywords":["refund","resolved","compensation","credit","waived"],"fuzzy":true,"metadata_filters":{"domain":"Hotel"},"approx_span_tokens":80,"temporal_anchor":"none","priority":"low"}
{"lines":6}

Query: What promo codes are most famous in retail domain?
-1: "dissatisfied-caller evidence about commonly mentioned retail promo codes" {"keywords":["promo","discount","code","offer","coupon"],"fuzzy":true,"metadata_filters":{"domain":"Retail"},"approx_span_tokens":120,"temporal_anchor":"none","priority":"high"}
0: "neutral-caller references listing retail promo codes frequently used" {"keywords":["promo","deal","sale","coupon","retail"],"fuzzy":true,"metadata_filters":{"domain":"Retail"},"approx_span_tokens":100,"temporal_anchor":"none","priority":"medium"}
1: "satisfied-caller mentions of popular retail promo codes" {"keywords":["promo","applied","savings","loyalty","coupon"],"fuzzy":true,"metadata_filters":{"domain":"Retail"},"approx_span_tokens":80,"temporal_anchor":"none","priority":"low"}
{"lines":6}

"""

def decompose_query(q: str, max_subqs: int = 6) -> list[str]:
    """
    Implements DECOMPOSE(q, p) from the paper.
    Returns a list of sub-queries ˜q1, ..., ˜qn (n ≤ max_subqs).
    """
    user_prompt = f"Original question:\n{q}\n\nSub-questions:"
    raw = call_llm(DECOMPOSE_PROMPT, user_prompt, max_tokens=512)
    lines = [l.strip() for l in raw.split("\n") if l.strip()]
    # Handle the {"lines":6} marker
    if lines and lines[-1] == '{"lines":6}':
        lines = lines[:-1]
    # Truncate to max_subqs
    subqs = lines[:max_subqs]
    # Filter trivial duplicates / exact copies of q
    subqs = [s for s in subqs if s.lower() != q.lower()]
    return subqs
```

```python

# -----------------------------
# Dense retrieval (TOP-K)
# -----------------------------
def dense_retrieve(query: str, k: int = 10) -> list[dict]:
    """
    Returns a list of candidate dicts:
    {"doc_id": str, "text": str, "score": float}
    """
    q_emb = embed_model.encode([query], convert_to_numpy=True, normalize_embeddings=True)
    scores, idxs = index.search(q_emb, k)
    results = []
    for score, idx in zip(scores[0], idxs[0]):
        p = passages_data[idx]
        results.append(
            {
                "doc_id": p["id"],
                "text": p["text"],
                "score": float(score),
            }
        )
    return results


# -----------------------------
# Reranker module (RERANK)
# -----------------------------
def rerank_candidates(original_query: str, candidates: list[dict]) -> list[dict]:
    """
    Given original query q and candidate passages_data C,
    returns candidates sorted by cross-encoder relevance score g_phi(q, d).
    """
    if not candidates:
        return []

    pairs = [(original_query, c["text"]) for c in candidates]
    scores = reranker.predict(pairs)  # shape: [len(candidates)]

    for c, s in zip(candidates, scores):
        c["rerank_score"] = float(s)

    candidates_sorted = sorted(candidates, key=lambda x: x["rerank_score"], reverse=True)
    return candidates_sorted

```

```python
# -----------------------------
# Generation module
# -----------------------------
GENERATION_SYSTEM_PROMPT = """You are a helpful assistant that answers questions
based only on the provided context passages_data. If the context is insufficient,
say that you cannot answer with certainty."""

def generate_answer(original_query: str, top_passages: list[dict], max_tokens: int = 256) -> str:
    """
    Concatenate q and top-k passages_data, then call a generator LLM.
    """
    context_blocks = []
    for i, p in enumerate(top_passages):
        context_blocks.append(f"[PASSAGE {i+1}]\n{p['text']}")
    context = "\n\n".join(context_blocks)

    user_prompt = (
        f"Question:\n{original_query}\n\n"
        f"Context passages_data:\n{context}\n\n"
        "Answer:"
    )
    answer = call_llm(GENERATION_SYSTEM_PROMPT, user_prompt, max_tokens=max_tokens)
    return answer.strip()
```

```python

# -----------------------------
# Full QD + RR pipeline (Algorithm 1)
# -----------------------------
def retrieve_with_qd_rag(q: str, k_retrieval: int = 10, k_final: int = 10) -> list[dict]:
    """
    Implements Algorithm 1 in the paper:
    1: Q ← {q} ∪ DECOMPOSE(q)
    2–5: for all q' in Q: C ← C ∪ TOP-K(q', D)
    6: C ← RERANK(C)
    7: R'(q) ← HEAD(C, k_final)
    """
    # Step 1: build query set Q
    subqs = decompose_query(q)  # {˜q1, ..., ˜qn}
    Q = [q] + subqs

    # Steps 2–5: retrieve for each query and build global candidate set C
    candidates_by_id = {}  # deduplicate by doc_id

    for q_i in Q:
        retrieved = dense_retrieve(q_i, k=k_retrieval)
        for cand in retrieved:
            doc_id = cand["doc_id"]
            # keep the best dense score for duplicates
            if doc_id not in candidates_by_id or cand["score"] > candidates_by_id[doc_id]["score"]:
                candidates_by_id[doc_id] = cand

    candidates = list(candidates_by_id.values())

    # Step 6: rerank against original complex query
    reranked = rerank_candidates(q, candidates)

    # Step 7: HEAD(C, k_final)
    top_k = reranked[:k_final]
    return top_k


def qd_rag_answer(q: str, k_retrieval: int = 10, k_final: int = 10, max_gen_tokens: int = 256) -> tuple[str, list[dict]]:
    """
    Convenience wrapper: runs full QD+RAG and returns (answer, top_passages).
    """
    top_passages = retrieve_with_qd_rag(q, k_retrieval=k_retrieval, k_final=k_final)
    answer = generate_answer(q, top_passages, max_tokens=max_gen_tokens)
    return answer, top_passages


if __name__ == "__main__":
    question = "Which product is being returned the most in retail?"
    answer, ctx = qd_rag_answer(question)
    print("QUESTION:", question)
    print("Decomposed Qs", decompose_query(question))
    print("Candidates sorted", candidates_sorted)
    print("Reranked", reranked)
    print("reranked selected", top_k)
    print("ANSWER:", answer)
    print("\nUSED PASSAGES:")
    for i, p in enumerate(ctx):
        print(f"{i+1}. {p['doc_id']} (rerank_score={p.get('rerank_score', None):.4f})")
        print("   ", p["text"])
```
